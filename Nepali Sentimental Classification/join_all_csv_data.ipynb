```python
import pandas as pd
```

```python
not_null_sentimental_data_path = "/content/drive/MyDrive/Colab Notebooks/NLP/Classification/Nepali Sentimental Classification/sentiment_data/not_null_sentimental_data.csv"
train_data_path = "/content/drive/MyDrive/Colab Notebooks/NLP/Classification/Nepali Sentimental Classification/sentiment_data/train.csv"
test_data_path = "/content/drive/MyDrive/Colab Notebooks/NLP/Classification/Nepali Sentimental Classification/sentiment_data/test.csv"

not_null_sentimental_data = pd.read_csv(not_null_sentimental_data_path)
train_data = pd.read_csv(train_data_path)
test_data = pd.read_csv(test_data_path)
```

# Non Null data preprocessing

```python
not_null_sentimental_data.drop(labels = ["Unnamed: 0"], axis = 1, inplace = True)
not_null_sentimental_data.head()
```

```python
not_null_sentimental_data.info()
```

# Train data preprocessing

```python
train_data.head()
```

```python
test_data.head()
```

```python
not_null_sentimental_data.nunique()
```

```python
not_null_sentimental_data["label"] = not_null_sentimental_data["label"].astype(int)
```

```python
not_null_sentimental_data.nunique()
```

```python
train_data.info()
```

```python
train_data["label"].value_counts()
```

```python
train_data.drop(train_data[train_data["label"] == "-"].index, axis = 0, inplace = True)
train_data.drop(train_data[train_data["label"] == "o"].index, axis = 0, inplace = True)
train_data.drop(train_data[train_data["label"] == "--"].index, axis = 0, inplace = True)
train_data.drop(train_data[train_data["label"] == "20"].index, axis = 0, inplace = True)
train_data.drop(train_data[train_data["label"] == "11"].index, axis = 0, inplace = True)
```

```python
train_data["label"].value_counts()
```

```python
train_data.info()
```

# test_data preprocessing

```python
test_data.head()
```

```python
test_data.info()
```

```python
test_data["label"].value_counts()
```

```python
test_data.drop(test_data[test_data["label"] == "-"].index, axis = 0, inplace = True)
test_data.drop(test_data[test_data["label"] == "o"].index, axis = 0, inplace = True)
```

```python
test_data["label"].value_counts()
```

```python
test_data.info()
```

# Concatenate all data

```python
concat_data = pd.concat([not_null_sentimental_data, train_data, test_data], axis = 0)
```

```python
concat_data.info()
```
