```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import joblib
```

```python
final_sentences = joblib.load("/content/drive/MyDrive/Colab Notebooks/NLP/Language Modelling/Datasets/Nepali_Corpus/final_sentences.pkl")
```

```python
final_sentences[0:5]
```

```python
word_cloud = ""
for row in final_sentences:
    for word in row:
        word_cloud+=" ".join(word)
```

```python
from wordcloud import WordCloud
```

```python
plt.rcParams["figure.figsize"] = (20,8)
font1 = "/content/drive/MyDrive/Colab Notebooks/NLP/Language Modelling/Word Embeddings/TiroDevanagariHindi-Regular.ttf"
wordcloud1 = WordCloud(width = 1000, height = 500, background_color = "white", min_font_size = 10, regexp=r"[\u0900-\u097F]+", font_path = font1).generate(word_cloud)
plt.imshow(wordcloud1)
```
