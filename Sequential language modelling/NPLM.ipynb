```python
oscar_dedup_file_path = "/content/drive/MyDrive/Colab Notebooks/NLP/Language Modelling/Datasets/Oscar_nepali_dataset/ne_dedup.txt"
```

```python
with open(oscar_dedup_file_path) as f:
  lines = f.readlines()
```

```python
oscar_paras = []
for sentence in tqdm(lines):
  words = sentence.split()
  oscar_paras.append(words) 
```

```python
train_size = int(len(oscar_paras) * 0.95)
train_size
```

```python
UNK_SYMBOL = "<UNK>"
vocab = set([UNK_SYMBOL])
vocab
```

```python
len(oscar_paras[6993])
```

```python
# create the corpus from training dataset
corpus = []
for idx, paragraph in enumerate(oscar_paras):
  if idx == train_size:
    break
  words = []
  for word in paragraph:
    words.append(word)
  corpus.append(words)
```

```python
words_term_frequency_train = {}
for sentence in tqdm(corpus):
  for word in sentence:
    words_term_frequency_train[word] = words_term_frequency_train.get(word, 0) + 1
```
